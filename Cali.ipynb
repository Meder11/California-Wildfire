```python
import pandas as pd
```

```python
# read the data with pandas
cali = pd.read_csv("Cali.csv")

# Print the first 5 rows of the data
cali.head()
```

```python
# Display the data types and null values

print(cali.info())
```

```python
# Check if there are duplicates and drop if found.

# Check if there are duplicates and drop if found.
duplicates = cali.duplicated().sum()  # Count duplicates

if duplicates > 0:
    print(f'Total duplicates: {duplicates}')
    cali = cali.drop_duplicates()  # Drop duplicates
else:
    print('No duplicates found')
print('No duplicates found')
```

```python
# Rename the column that are too long
cali = cali.rename(columns={"Area_Burned (Acres)": "Area_Burned",
                            "Estimated_Financial_Loss (Million $)": "Estimated_Loss"})
print(cali.head())
```

```python
# Describe the data
cali.describe()
```

```python
# Get the last rows of the data.
cali.tail()
```

```python
# Change date to date time
cali["Date"] = pd.to_datetime(cali["Date"])
```

```python
# Trim whitespace
cali.columns = cali.columns.str.strip()
cali["Cause"] = cali["Cause"].str.strip()
```

```python
# Lowercase text
cali["Cause"] = cali["Cause"].str.lower()
```

```python
# Ensure categorical values are consistent
print(cali["Cause"].unique())  # Check unique values
```

```python
cali.to_csv("califo.csv")
print("File is saved")
```
